```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
size = 365*8

newweighting = (300000/size-5000*np.cos(np.pi*(np.arange(size)+200)/size*2)/size*2*np.pi 
                    +2000*np.cos(np.pi*(np.arange(size)+200)/size*4)/size*4*np.pi)/100.

take2 =  1 + 0.75 * (newweighting - 1)
```

```python
plt.plot(newweighting)
plt.plot(take2);
print (newweighting.mean(), take2.mean())
```

```python
weight = np.zeros((12))
monthlength = [31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]
start = 0
for im in range(12):
    end = start + monthlength[im]
    weight[im] = newweighting[start*8:end*8].mean()
    start = end
```

```python
plt.plot(weight)
```
