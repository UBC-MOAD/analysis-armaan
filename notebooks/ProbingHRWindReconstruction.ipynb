```python
import matplotlib.pyplot as plt
import numpy as np
import xarray as xr
```

```python
orig = xr.open_dataset('ncfiles/vectoravg_y2007m11d12.nc')
HF = xr.open_dataset('ncfiles/recon_HF_y2007m11d12.nc')
```

```python
fig, axs = plt.subplots(1, 2, figsize=(12, 5))
iday = 5
orig.v_wind[iday].plot(ax=axs[0]);
HF.v_wind[iday].plot(ax=axs[1]);
for ax in axs:
    ax.plot(160, 120, 'x')
```

```python
fig, axs = plt.subplots(1, 3, figsize=(15, 5))
orig.v_wind[:, 120, 160].plot(ax=axs[0]);
HF.v_wind[:, 120, 160].plot(ax=axs[0]);
orig.u_wind[:, 120, 160].plot(ax=axs[1]);
HF.u_wind[:, 120, 160].plot(ax=axs[1]);
np.sqrt(orig.u_wind[:, 120, 160]**2 + orig.v_wind[:, 120, 160]**2).plot(ax=axs[2]);
np.sqrt(HF.u_wind[:, 120, 160]**2 + HF.v_wind[:, 120, 160]**2).plot(ax=axs[2]);
```

```python
orig.close()
HF.close()
```
